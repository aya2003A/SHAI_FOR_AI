```python
import pandas as pd
import numpy as np

df = pd.read_csv('Salaries.csv')
df.head()
```

```python
print("number of rows ",df.shape[0])
print("number of col ",df.shape[1])
```

```python
df.dtypes
```

```python
df.isnull()
```

```python
print(" mean ",df['TotalPay'].mean())
print(" median ",df['TotalPay'].median())
print(" mode ",df['TotalPay'].mode())
```

```python
print(" min ",df['TotalPay'].min())
print(" max ",df['TotalPay'].max())
print(" std ",df['TotalPay'].std())
print("range",df['TotalPay'].max()-df['TotalPay'].min())
```

```python
#dropped column that all of it is values are null
df.dropna(axis=1 ,how="all", inplace=True)
```

```python
df.dropna(axis=0 ,how="all", inplace=True)
```

```python
df['BasePay']=df['BasePay'].fillna(0)
df['OvertimePay']=df['OvertimePay'].fillna(0)
df['OtherPay']=df['OtherPay'].fillna(0)
df['Benefits']=df['Benefits'].fillna(0)
#replace all null values in columns that have null values on it
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
df.hist(column="TotalPay",color='#4B7BCC',edgecolor = "black")
plt.title("distribution of salaries")
plt.xlabel("Range")
plt.ylabel("values")
plt.show()
```

```python
df[['JobTitle','Departments']]=df['JobTitle'].str.split('(',expand=True)
df
```

```python
df.Departments.value_counts()
```

```python
state_by_year=df.groupby("Year")
```
